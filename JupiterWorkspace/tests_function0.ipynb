```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
```

```python
plt.plot(x, y)      # 创建线图
plt.xlabel('X Axis Label')  # X轴标签
plt.ylabel('Y Axis Label')  # Y轴标签
plt.title('Title of Plot')  # 图表标题
plt.show()           # 显示图表
```

```python
import math
import numpy as np

for i in range(0, 10):
    n = math.pow(2, i)
    a = np.arange(-10.0, 10.0, 0.1, dtype=np.longdouble)
    try:
        b = [(1.0 / (1.0 + n * math.pow(x, n))) for x in a]
        plt.plot(a, b)
        plt.xlabel("x")
        plt.ylabel("2^%d" % i)
        plt.show()
    except Exception as ex:
        print(ex)
```
